```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
starting_rf = RandomForestClassifier(random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
df.drop(["artist", "track", "uri"], axis=1, inplace=True)
y = df["grupa"]
X = df.drop("grupa", axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import f1_score, roc_curve, auc, confusion_matrix, ConfusionMatrixDisplay

accuracy_scores = []
f1_scores = []
roc_scores = {}
auc_scores = {} 

starting_rf.fit(X_train, y_train)
y_hat_train = starting_rf.predict(X_train)
y_hat_test = starting_rf.predict(X_test)

accuracy_train = starting_rf.score(X_train, y_train)
accuracy_test = starting_rf.score(X_test, y_test)
accuracy_scores.append(["Las losowy", accuracy_train, "Ciag uczacy"])
accuracy_scores.append(["Las losowy", accuracy_test, "Ciag testowy"])

f1_train = f1_score(y_train, y_hat_train)
f1_test = f1_score(y_test, y_hat_test)
f1_scores.append(["Las losowy", f1_train, "Ciag uczacy"])
f1_scores.append(["Las losowy", f1_test, "Ciag testowy"])

fpr, tpr, threshold = roc_curve(y_test, y_hat_test)
test_auc = auc(fpr, tpr)

roc_scores["Las losowy"] = [fpr, tpr]
auc_scores["Las losowy"] = test_auc

print(accuracy_train)
print(accuracy_test)

cm=confusion_matrix(y_test, y_hat_test)
display = ConfusionMatrixDisplay(confusion_matrix=cm)
display.plot()
plt.show()

ax= plt.subplot()
sns.heatmap(cm, annot=True, fmt='g', ax=ax);
# labels, title and ticks
ax.set_xlabel('Przypisana klasa');ax.set_ylabel('Rzeczywista klasa'); 
ax.xaxis.set_ticklabels(['kit', 'hit']); ax.yaxis.set_ticklabels(['kit', 'hit']);
```
